```python
#Start a loop so the user is allowed to input a number
while True:
    try:
        user_input = input("Enter a number (or type 'exit' to quit): ")  #Has user input chosen number (sting)
        if user_input.lower() == 'exit':                                #User enters 'exit' to end
            break
        number = int(user_input)                                         #The int changes the users input from the string to an integer
        if number % 3 == 0 and number % 5 == 0:                           #Checks if the # is divisible by 3 and 5
            print("FizzBuzz")
        elif number % 3 == 0:                                             #Checks if divisible by 3 only
            print("Fizz")
        elif number % 5 == 0:                                              #Checks if divisible by 5 only
            print("Buzz")
        else:                                                              #If the # is not divisible by 3 or 5 the loops ends
            print(number)
    except ValueError:
        print("Invalid input. Please enter an integer.")
```
